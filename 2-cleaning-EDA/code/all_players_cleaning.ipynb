```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../1-data-collection/data/all_replays.csv')
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.player_rank = df.player_rank.str.replace(' III ', ' 3 ')
df.player_rank = df.player_rank.str.replace(' II ', ' 2 ')
df.player_rank = df.player_rank.str.replace(' I ', ' 1 ')
```

```python
df['numeric_rank'] = 0
```

```python
def numeric_rank(df, rank_col):
    df['numeric_rank'] = 0
    rank_dict = {
        'Bronze' : 0,
        'Silver' : 1,
        'Gold' : 2,
        'Platinum' : 3,
        'Diamond' : 4,
        'Champion' : 5,
        'Grand' : 6,
        'Supersonic' : 7
    }
    for i in range(len(rank_col)):
        if rank_col[i] == 'Supersonic Legend':
            df.numeric_rank[i] = 7 * 12 + 1
        if rank_col[i] == 'Unranked':
            df.numeric_rank[i] = np.nan
        try:
            
            df['numeric_rank'][i] = int(rank_col[i].split(' ')[-1]) + ((int(rank_col[i].split(' ')[-3]) -1) * 4) + (rank_dict[rank_col[i].split(' ')[0]] * 12)
        
        except ValueError:
            
            pass
        
    return df
```

```python
numeric_rank(df, df.player_rank)
```
