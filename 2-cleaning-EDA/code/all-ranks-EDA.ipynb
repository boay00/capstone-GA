## Rank EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/cleaned_random_replays.csv')
```

```python
df.columns
```

```python
sns.set_style('dark')
plt.plot(df.groupby('numeric_rank')['bcpm'].agg(lambda x: x.mean()), color = 'darkblue')
plt.xlabel('Numeric Rank')
plt.ylabel('Average Boost Consumed per Minute')
plt.title('Boost Consumption at Each Rank')
plt.axvline(1.5, alpha = 0.4, color = 'sienna')
plt.text(2, 275, '''Bronze''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(13, alpha = 0.4, color = 'silver')
plt.text(13.5, 305, '''Silver''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(25, alpha = 0.4, color = 'gold')
plt.text(25.5, 325, '''Gold''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(37, alpha = 0.4, color = 'cyan')
plt.text(37.5, 360, '''Platinum''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(49, alpha = 0.4, color = 'blue')
plt.text(49.5, 385, '''Diamond''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(61, alpha = 0.4, color = 'orchid')
plt.text(61.5, 405, '''Champion''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(73, alpha = 0.4, color = 'red')
plt.text(73.5, 435, '''Grand
Champion''', fontsize=7, zorder = 11, alpha = 0.5)
plt.axvline(84, alpha = 1, color = 'white')
plt.text(84.5, 445, '''SSL''', fontsize=7, zorder = -1, alpha = 0.5)
```

- As ranks progress towards Supersonic Legend, the boost consumed increases linearly
- SSL has lower boost consumption than Grand Champion, however very few replays were sampled from SSL players, so this is not a reliable finding

- Avg powerslide duration increases until silver ranks, then decreases to around 0.10 seconds.
- 0.10 seconds is the sweetspot, in which car momentum will be retained without being lost, and turning will be maximised from powerslide
